**17.1 Interpolation Problem Statement**

*   Interpolation is estimating the function using the data given
*   Compared to regression, interpolation doesn't require an underlying model for the data. However, the underlying process is still needed to assess the quality of interpolation.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [1,2,3,4,6,8,10]
y = [1,2.5,3,4,5,4,3]

plt.figure(figsize = (10,8))
plt.plot(x,y,'bo', 5,4.5,'ro')
plt.title('Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.2 Linear Interpolation**

*   In linear interpolation, the estimated point (in red) is assumed to lie on the line joining the nearest points to the left and right.
*   The equation for this point is:

y(x) = yi + [(y_(i+1) - y_i)(x - x_i)] / [x_(i+1) - x_i]

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [1,2,3,4,6,8,10]
y = [1,2.5,3,4,5,4,3]

f = interp1d(x, y)
y_hat = f(5)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x,y,'-ob')
plt.plot(5,y_hat,'ro')
plt.title('Linear Interpolation at x = 5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.3 Cubic Spline Interpolation**

*   The cubic interpolating function is a set of piecewise cubic functions
*   For n points, there are n-1 cubic functions to find. Each function has its own 4 coefficients that are joined by a cubic polynomial.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,4,6,8,10]
y = [1,2.5,3,4,5,4,3]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 10, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
